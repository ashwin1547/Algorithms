```python
def getWays(n, c):
    # Write your code here
    n_ways = [0]*(n+1)
    for i in range(n+1):
        n_ways[i] = [0]*len(c)
    for i in range(len(c)):
        n_ways[0][i] = 1
    for i in range(1,n+1):
        for j in range(len(c)):
            x = n_ways[i - c[j]][j] if i-c[j] >= 0 else 0
  
            # Count of solutions excluding S[j] 
            y = n_ways[i][j-1] if j >= 1 else 0
  
            # total count 
            n_ways[i][j] = x + y 
  
        
    return n_ways[n][len(c)-1]
getWays(3,[3,1,2])
```

# MergeSort

```python
def Merge(B,C):
    D = []
    while len(B)!=0 and len(C)!=0:
        b = B[0]
        c = C[0]
        if (b<=c):
            D.append(b)
            B.remove(b)
        else:
            D.append(c)
            C.remove(c)
    if (len(B)!= 0):
        for i in range(len(B)):
            D.append(B[i])
    if (len(C)!= 0):
        for i in range(len(C)):
            D.append(C[i])
    return D

def MergeSort(A):
    if (len(A)== 1):
        return A
    B = MergeSort(A[:len(A)//2])
    C = MergeSort(A[len(A)//2:])
    A = Merge(B,C)
    return A
```

```python
A = [10,9,8,7,6,5]
r = len(A)-1
l = 0
def QuickSort(A,l,r):
    if l>=r:
        return A
    m = Partition(A,l,r)
    QuickSort(A,l,m-1)
    QuickSort(A,m+1,r)
    
def Partition(A,l,r):
    x = A[l]
    j = l
    for i in range(l+1,r+1):
        if A[i]<=x:
            j = j+1
            A[j],A[i] = A[i],A[j]
    A[l],A[j] = A[j],A[l]
    return j
```

# No of ways to climb stairs with stairs n and max step m

```python
def No_of_ways(n,m):
    ways = [0]*(n+1)
    ways[0] = 1
    for i in range(1,n+1):
        x = 0
        for j in range(1,m+1):
            if n>=j:
                x = x + ways[i-j]
        ways[i] = x
    return ways
No_of_ways(5,3)
```

# Collecting Maximum value items in a square maze while reaching from top left corner to right left corner with given valid moves
Moves in our case: right or down

```python
l = [[1,0,5],[2,10,2],[0,1,1]]
dp = [0]*len(l)
for i in range(len(l)):
    dp[i] = [0]*len(l[0])
dp[0][0] = l[0][0]
for i in range(1,len(l)):
    dp[i][0] = dp[i-1][0]+l[i][0]
    dp[0][i] = dp[0][i-1]+l[0][i]
for i in range(1,len(l)):
    for j in range(1,len(l)):
        dp[i][j] = max(dp[i-1][j]+l[i][j],dp[i][j-1]+l[i][j])
print(dp[len(l)-1][len(l)-1])
```
